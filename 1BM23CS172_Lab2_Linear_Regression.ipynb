```python
# -*- coding: utf-8 -*-
"""Linear-Regression-Housing_Area_Price.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1CAlZml-P6V2V1RIrodgMfF8L3Ux4V9FT
"""

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('/content/housing_area_price.csv')
df

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df

price = df.price
price

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)

"""(1) Predict price of a home with area = 3300 sqr ft"""

reg.predict([[3300]])

reg.coef_

reg.intercept_

"""Y = m * X + b (m is coefficient and b is intercept)"""

3300*135.78767123 + 180616.43835616432

"""(1) Predict price of a home with area = 5000 sqr ft"""

reg.predict([[5000]])
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Load the dataset
df = pd.read_csv("/content/canada_per_capita_income.csv")

# Display first few rows
print(df.head())

# Step 3: Data Preprocessing
# Check for null values
print(df.isnull().sum())

# If null values exist, drop them
df = df.dropna()

# Define independent (X) and dependent (y) variables
X = df[['year']]   # Independent variable must be 2D
y = df['per capita income (US$)']

# Step 4: Create and train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 5: Predict income for year 2020
year_2020 = np.array([[2020]])
predicted_income = model.predict(year_2020)

print("Predicted Per Capita Income for Canada in 2020: $", predicted_income[0])

# Step 6 (Optional): Visualization
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(2020, predicted_income[0], color='green', label='Prediction (2020)')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.show()
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Load dataset
df = pd.read_csv("/content/salary.csv")

# Display first few rows
print(df.head())

# Step 3: Data Preprocessing

# Check column names
print(df.columns)

# Check for missing values
print(df.isnull().sum())

# Drop missing values if any
df = df.dropna()

# Define independent and dependent variables
X = df[['YearsExperience']]   # Independent variable (must be 2D)
y = df['Salary']              # Dependent variable

# Step 4: Create Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Step 5: Predict salary for 12 years of experience
experience_12 = np.array([[12]])
predicted_salary = model.predict(experience_12)

print("Predicted Salary for 12 years of experience: ₹", predicted_salary[0])

# Step 6 (Optional): Visualization
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(12, predicted_salary[0], color='green', label='Prediction (12 years)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 2: Load dataset
df = pd.read_csv("/content/hiring.csv")

# Display dataset
print(df.head())

# Step 3: Data Preprocessing

# Check for missing values
print(df.isnull().sum())

# Some hiring datasets contain 'experience' in words like:
# zero, one, two, three, etc.
# Convert words to numbers if needed

word_to_num = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
}

if df['experience'].dtype == object:
    df['experience'] = df['experience'].map(word_to_num)

# Fill missing experience with 0
df['experience'] = df['experience'].fillna(0)

# Fill missing test scores with mean value
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mean()
)

# Step 4: Define independent and dependent variables
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Step 5: Create and train Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 6: Make Predictions

# Candidate 1: 2 yr experience, 9 test score, 6 interview score
salary1 = model.predict([[2, 9, 6]])

# Candidate 2: 12 yr experience, 10 test score, 10 interview score
salary2 = model.predict([[12, 10, 10]])

print("Predicted Salary for Candidate 1: $", salary1[0])
print("Predicted Salary for Candidate 2: $", salary2[0])
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 2: Load dataset
df = pd.read_csv("/content/1000_Companies.csv")

# Display first few rows
print(df.head())

# Step 3: Data Preprocessing

# Check for missing values
print(df.isnull().sum())

# Drop missing values if any
df = df.dropna()

# Separate features and target
X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y = df['Profit']

# Step 4: Encode categorical variable 'State'
X = pd.get_dummies(X, columns=['State'], drop_first=True)

# Step 5: Train Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 6: Prepare input for prediction

# Create a dataframe with same structure as training data
input_data = pd.DataFrame({
    'R&D Spend': [91694.48],
    'Administration': [515841.3],
    'Marketing Spend': [11931.24],
    'State': ['Florida']
})

# Apply same encoding
input_data = pd.get_dummies(input_data, columns=['State'], drop_first=True)

# Ensure all columns match training data
input_data = input_data.reindex(columns=X.columns, fill_value=0)

# Step 7: Predict profit
predicted_profit = model.predict(input_data)

print("Predicted Profit: $", predicted_profit[0])
```
